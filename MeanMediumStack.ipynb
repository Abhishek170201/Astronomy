```python
# Write your mean_fits function here:
from astropy.io import fits
def mean_fits(imglist):
  n=len(imglist)
  hdulist=fits.open(imglist[0])
  data=hdulist[0].data
  for i in range(1,n):
    hdulist=fits.open(imglist[i])
    data+=hdulist[0].data
  data/=n
  return data



if __name__ == '__main__':
  
  # Test your function with examples from the question
  data  = mean_fits(['image0.fits', 'image1.fits', 'image2.fits'])
  print(data[100, 100])

  # You can also plot the result:
  import matplotlib.pyplot as plt
  plt.imshow(data.T, cmap=plt.cm.viridis)
  plt.colorbar()
  plt.show()
```

```python
# Write your function median_FITS here:
from astropy.io import fits
import time
import numpy as np
def median_fits(imglist):
  n=len(imglist)
  big_data=[]
  start=time.perf_counter()
  for i in range(n):
    hdulist=fits.open(imglist[i])
    data=hdulist[0].data
    big_data.append(data)
    hdulist.close()
  FITS_stack=np.dstack(big_data)
  median=np.median(FITS_stack,axis=2)
  timetaken=time.perf_counter()-start
  memory=FITS_stack.nbytes/1024
  return (median,timetaken,memory)

# You can use this to test your function.
# Any code inside this `if` statement will be ignored by the automarker.
if __name__ == '__main__':
  # Run your function with first example in the question.
  result = median_fits(['image0.fits', 'image1.fits'])
  print(result[0][100, 100], result[1], result[2])
  
  # Run your function with second example in the question.
  result = median_fits(['image{}.fits'.format(str(i)) for i in range(11)])
  print(result[0][100, 100], result[1], result[2])
```
